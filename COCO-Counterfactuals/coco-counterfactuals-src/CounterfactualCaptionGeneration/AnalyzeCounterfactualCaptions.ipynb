```python
import nltk
p = nltk.PorterStemmer()
import re
```

```python
regex = re.compile('[^a-zA-Z ]')
```

```python
prompt_file = './output/gen_captions/perplexity_score/sim_.8_.91/cap_val_2017.txt'
origin_caption = {}
counter_caption = {}
with open(prompt_file, 'r') as infile:
    for l in infile:
        if l[-1] == '\n':
            l = l[:-1]
        index, origin, counter = l.split('\t')
        origin_caption[index] = origin
        counter_caption[index] = counter
```

```python
print(f'len of counterfactual captions: {len(counter_caption)}')
```

```python
def extract_replaced_words(original, counter):
    original = regex.sub('', original)
    counter = regex.sub('', counter)
    
    original=original.split()
    # print(original)
    counter=counter.split()
    # print(counter)
    replaced_word = None
    being_replaced_word = None
    i, j = -1, -1
    for i, w in enumerate(original):
        if w not in counter:
            being_replaced_word = w
            break
    for j, w in enumerate(counter):
        if w not in original:
            replaced_word = w
            break
    if replaced_word == None and being_replaced_word != None:
        replaced_word = original[j]
    if being_replaced_word == None and replaced_word != None:
        being_replaced_word = counter[i]
    if being_replaced_word == None and replaced_word == None:
        print('Error', original, counter)
    return being_replaced_word, replaced_word
```

```python
changed_words = {}
replacing_words = {}
stem_word_map = {}
changed_word_info = []
for index in origin_caption.keys():
    being_replaced_word, replaced_word = extract_replaced_words(origin_caption[index], counter_caption[index])
    if being_replaced_word != None and replaced_word != None:
        stem_word_map[p.stem(being_replaced_word)] = being_replaced_word
        stem_word_map[p.stem(replaced_word)] = replaced_word
        changed_words[index] = {'origin_word' : being_replaced_word, 
                                'counterfactual_word' : replaced_word,
                                'origin_stem_word' : p.stem(being_replaced_word),
                                'counterfactual_stem_word' : p.stem(replaced_word),
                               }
        changed_word_info.append({'origin_word' : being_replaced_word, 
                                'counterfactual_word' : replaced_word,
                                'origin_stem_word' : p.stem(being_replaced_word),
                                'counterfactual_stem_word' : p.stem(replaced_word),
                                  'index' : index,
                               })
```
